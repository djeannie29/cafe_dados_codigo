# **Conectando SQLite no Google Colab**


# **Importações:**

```python
import sqlite3
import pandas as pd
```

# **Criar conexão (banco em memória)**

```python
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
```

# **Criar tabela e inserir dados**

```python
cursor.execute('CREATE TABLE alunos (id INTEGER, nome TEXT, nota REAL)')
cursor.executemany('INSERT INTO alunos VALUES (?, ?, ?)', [
    (1, 'Ana', 8.5),
    (2, 'João', 7.0),
    (3, 'Maria', 9.2)
])
conn.commit()
```

# **Consultar com SQL direto no Python**

```python
df = pd.read_sql_query('SELECT * FROM alunos', conn)
print("=== Tabela de Alunos ===")
display(df)
```

# **Exemplo**

```python
df_filtrado = pd.read_sql_query('SELECT nome, nota FROM alunos WHERE nota >= 8', conn)
print("\n=== Alunos com nota ≥ 8 ===")
display(df_filtrado)
```
